```python

!pip install psutil tabulate colorama

import psutil
import socket
import platform
from datetime import datetime
from tabulate import tabulate
from colorama import Fore, Style, init


init(autoreset=True)


def system_info():
    print(Fore.CYAN + "\n========== SYSTEM INFORMATION ==========")

    hostname = socket.gethostname()

    try:
        ip_address = socket.gethostbyname(hostname)
    except:
        ip_address = "127.0.0.1"

    info = [
        ["Hostname", hostname],
        ["IP Address", ip_address],
        ["Operating System", platform.system()],
        ["OS Version", platform.version()],
        ["Machine", platform.machine()],
        ["Processor", platform.processor()],
        ["Current Time", datetime.now().strftime("%Y-%m-%d %H:%M:%S")]
    ]

    print(tabulate(info, tablefmt="grid"))


def resource_usage():
    print(Fore.CYAN + "\n========== RESOURCE USAGE ==========")

    cpu = psutil.cpu_percent(interval=1)
    memory = psutil.virtual_memory()

    data = [
        ["CPU Usage", f"{cpu}%"],
        ["Memory Usage", f"{memory.percent}%"],
        ["Available Memory", f"{round(memory.available / (1024**3), 2)} GB"]
    ]

    print(tabulate(data, tablefmt="grid"))


def running_processes():
    print(Fore.CYAN + "\n========== RUNNING PROCESSES ==========")

    processes = []

    for proc in psutil.process_iter(['pid', 'name', 'username']):
        try:
            processes.append([
                proc.info['pid'],
                proc.info['name'],
                proc.info['username']
            ])
        except:
            pass

    print(tabulate(
        processes[:15],
        headers=["PID", "Process", "User"],
        tablefmt="grid"
    ))



def network_connections():
    print(Fore.CYAN + "\n========== NETWORK CONNECTIONS ==========")

    connections = []

    suspicious_ports = [21, 23, 4444, 5555, 1337]

    for conn in psutil.net_connections(kind='inet'):
        try:
            local_addr = f"{conn.laddr.ip}:{conn.laddr.port}"

            if conn.raddr:
                remote_addr = f"{conn.raddr.ip}:{conn.raddr.port}"
            else:
                remote_addr = "-"

            status = conn.status

            alert = ""

            if conn.laddr.port in suspicious_ports:
                alert = "Suspicious Port"

            connections.append([
                local_addr,
                remote_addr,
                status,
                alert
            ])

        except:
            pass

    print(tabulate(
        connections[:20],
        headers=["Local Address", "Remote Address", "Status", "Alert"],
        tablefmt="grid"
    ))


def logged_users():
    print(Fore.CYAN + "\n========== LOGGED-IN USERS ==========")

    users = []

    for user in psutil.users():
        users.append([
            user.name,
            user.host,
            datetime.fromtimestamp(user.started).strftime('%Y-%m-%d %H:%M:%S')
        ])

    if users:
        print(tabulate(
            users,
            headers=["Username", "Host", "Login Time"],
            tablefmt="grid"
        ))
    else:
        print("No logged-in users found.")


def security_alerts():
    print(Fore.CYAN + "\n========== SECURITY ALERTS ==========")

    alerts = []

    for conn in psutil.net_connections(kind='inet'):
        try:
            if conn.laddr.port in [4444, 5555, 1337]:
                alerts.append([
                    conn.laddr.port,
                    conn.status,
                    "Possible suspicious listener"
                ])
        except:
            pass

    if alerts:
        print(Fore.RED + tabulate(
            alerts,
            headers=["Port", "Status", "Warning"],
            tablefmt="grid"
        ))
    else:
        print(Fore.GREEN + "No suspicious ports detected.")


def main():
    print(Fore.YELLOW + Style.BRIGHT)
    print("========================================")
    print(" CYBERSECURITY SYSTEM MONITOR TOOL")
    print("========================================")

    system_info()
    resource_usage()
    running_processes()
    network_connections()
    logged_users()
    security_alerts()

    print(Fore.GREEN + "\nMonitoring completed successfully.")


main()
```
